## Prep

```python
data = data.drop([27, 28],axis=1)
```

```python
from sklearn.model_selection import train_test_split

#Stratified split due to class imbalance
X_train, X_test, y_train, y_test = train_test_split(data.drop(29, axis=1), data[29]=='Outlier', stratify=(data[29]=='Outlier'), random_state=42)
```

```python
X, y = data.drop(29, axis=1), (data[29] == 'Outlier')
```

## Logistic regression

Supervised pipeline is simple:
Since data is already normilized, and prepared (no None values), we need to deal with class imbalance.
Supervised pipeline uses SMOTE for this purpose

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as pl
from sklearn.linear_model import LogisticRegression


smt = SMOTE(random_state=42)

# Add SMOTE to deal with class imbalance

model = LogisticRegression(solver = 'saga' ,penalty='elasticnet', random_state=42, class_weight='balanced', l1_ratio=0.5)
logistic_pipeline = pl(steps=[
    ('SMOTE', smt),
    ('model', model)
])
```

```python
%%time
logistic_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import precision_recall_curve, auc, roc_auc_score, f1_score, precision_score, recall_score
```

```python
# Requaired metrics for supervised part

def evaluate(y_true, y_pred):
    print('f1 ', f1_score(y_true, y_pred))
    print('precision', precision_score(y_true, y_pred))
    print('recall', recall_score(y_true, y_pred))
    
    print('roc-auc ', roc_auc_score(y_true, y_pred))

    precision, recall, thresholds = precision_recall_curve(y_true, y_pred)
    print('pr-auc', auc(precision, recall))
```

```python
evaluate(y_test, logistic_pipeline.predict(X_test))
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
import numpy as np

cv = StratifiedKFold(5, random_state=42)
tuning_param = {
    'model__l1_ratio': [0.05, 0.2, 0.35, 0.5, 0.6, 0.8, 1],
    'model__C': [1e-5, 1e-3, 1e-1, 1e1, 1e3],
    'model__max_iter': range(50, 200, 50),
   }

```

```python
clf = GridSearchCV(logistic_pipeline, cv=cv, scoring='roc_auc', param_grid=tuning_param)
clf.fit(X_train, y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
evaluate(y_test, clf.predict(X_test))
```

## SVM

```python
from sklearn.svm import LinearSVC

model = LinearSVC(class_weight='balanced', random_state=42)
SVM_pipeline = pl(steps=[
    ('SMOTE', smt),
    ('model', model)
])
```

```python
cv = StratifiedKFold(5, random_state=42)
tuning_param = {
    'model__C': [1e-4, 1e-3, 1e-2, 0.1, 10],
    'model__tol': [1e-4, 1e-3, 1e-2]
   }

clf = GridSearchCV(SVM_pipeline, cv=cv, scoring='roc_auc', param_grid=tuning_param)
clf.fit(X_train, y_train)
```

```python
evaluate(y_test, clf.predict(X_test))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_jobs=-1)
knn_pipeline = pl(steps=[
    ('SMOTE', smt),
    ('model', model)
])
```

```python
cv = StratifiedKFold(5, random_state=42)
tuning_param = {
    'model__n_neighbors': [3, 5, 7, 9, 10, 15],
    'model__weights': ['uniform', 'distance'],
    'model__leaf_size': [10, 30, 45, 60],
   }

clf = GridSearchCV(knn_pipeline, cv=cv, scoring='roc_auc', param_grid=tuning_param)
clf.fit(X_train, y_train)
```

```python
evaluate(y_test, clf.predict(X_test))
```

## Unsupervised part

Unsupervised pipeline doesn't deal with the class imbalance. That's why the "pipeline" consists of the model only

```python
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import normalized_mutual_info_score
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score

# Required metrics for unsupervised part
def evaluate_unsup(y_true, y_pred, X):
    print('ARI ', adjusted_rand_score(y_true, y_pred))
    print('NMI ', normalized_mutual_info_score(y_true, y_pred))
    print('Silhouette coef ', silhouette_score(X, y_pred))
    print('Calinski Harabasz index ', calinski_harabasz_score(X, y_pred))
```

### One class SVM

3 of the model could be used for new records prediction, DBSCAN can only be used to predict labels for train sample. That's why there is 2 evalutions. On test sample, and the whole one

```python
from sklearn.svm import OneClassSVM

model = OneClassSVM(verbose=True)
```

```python
tuning_param = {
    'gamma': ['scale', 'auto'],
    'tol': [1e-3, 1e-4, 1e-2]
   }
```

```python
from sklearn.model_selection import GridSearchCV
clst = GridSearchCV(model, scoring='adjusted_rand_score', param_grid=tuning_param)
```

```python
clst.fit(X_train, y_train)
```

```python
evaluate_unsup(y_test, clst.predict(X_test), X_test)
```

```python
clst.best_params_
```

```python
model = OneClassSVM(gamma='auto', tol=0.01)
evaluate_unsup(y, model.fit_predict(X), X)
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
tuning_param = {
    'eps': [0.1, 0.9, 2, 3, 5],
    'min_samples': [3, 4, 8]
   }
```

```python
%%time
for e in tuning_param['eps']:
    for m in tuning_param['min_samples']:
        print(e, m)
        model = DBSCAN(eps = e, min_samples=m, n_jobs=-1)
        model.fit(X_train)
        if len(set(model.labels_)) > 1:
            evaluate_unsup(y_train, model.labels_, X_train)
        else:
            print('No outliers')
```

```python
%%time
for e in tuning_param['eps']:
    for m in tuning_param['min_samples']:
        print(e, m)
        model = DBSCAN(eps = e, min_samples=m, n_jobs=-1, metric='cosine')
        model.fit(X_train)
        if len(set(model.labels_)) > 1:
            evaluate_unsup(y_train, model.labels_, X_train)
        else:
            print('No outliers')
```

```python
model = DBSCAN(eps = 0.9, min_samples=3, metric='cosine', n_jobs=-1)
evaluate_unsup(y, model.fit_predict(X), X)
```
